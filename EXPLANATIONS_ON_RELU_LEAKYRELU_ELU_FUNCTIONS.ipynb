# CLEAR EXPLANATIONS ON RELU, LEAKY RELU, AND ELU FUNCTIONS
###  -  DEFINITIONS
- __What are RELU, LEAKY RELU AND ELU FUNCTIONS?__
- __Importances of these functions__

Activation functions are important in many Artificial Intelligence algorithms. For instance, in Logistic Regression 

classification algorithm, the mouthpiece of the dynamics of the classifier is logistic function: 
                            
 \begin{equation}
 f(x) =
   \frac{1}{e{^-x}+{1}}
\end{equation}

The expression becomes approximately zero at sufficiently small negatives and one at sufficiently large positives. It is a very 

useful function especially for binary classification inference problems. However, it causes gradient vanishing during 

parameter optimization in deep neural networks and thus making its use in such cases discouraged.
    

```python
import numpy as np
import matplotlib.pyplot as plt
values=np.arange(-5,5,0.1)
logf=1/((np.exp(-values)+1))
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(values, logf, color='red')
ax.set_title('Logistic Function', color='green')
plt.show()
```

## RECTIFIED LINEAR UNIT FUNCTION (RELU)

RELU function, on the other end, is a very useful function in neural network training. It makes deep networks easy and fast 

to train because it causes deactivation of some neurons by setting their weight values at zero and thus allows sparse 

network for computational training. This function is extremely useful in datasets involving huge number of features, and 

it is this reason it is a popular activation function in Convolutional Neural Networks. Nevertheless, it reduces network 

performances in some applications especially when many input values are negative.

RELU function, mathematically, is represented as:


> # __f(x) = max(0,x)__

\begin{equation}
  {f(x)} = \begin{cases} x & \text{if $x >= 0 $}\\0 & \text{otherwise}\end{cases}
\end{equation}
    

```python
import numpy as np
import matplotlib.pyplot as plt
values = np.arange(-5,5,0.1)
def relu(x):
    return max(0.,x)
relu=np.vectorize(relu)
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(values,relu(values),color='red')
ax.set_title('RELU Function', color='green')
plt.show()
```

## LEAKY RECTIFIED LINEAR UNIT FUNCTION (LEAKY RELU)

Leaky Relu is a variant of RELU that does not return zero values for negative inputs rather a small value derived from 

scaled input. The scalar is a small factor  __'s'__  that reduces drastically the original  absolute values of the negative 

inputs depending on the value of s, but not to zero. It performs better than RELU for neural networks training. 

Mathematically, Leaky RELU is represented as:


> # __f(x) = max(sx,x)__

\begin{equation}
  {f(x)} =\begin{cases} x & \text{if $x >= 0 $}\\sx & \text{otherwise; where s is a small positive value scaler}\end{cases}
\end{equation}
    

```python
import numpy as np
import matplotlib.pyplot as plt
values = np.arange(-5,5,0.5)
s=0.1
def lrelu(x):
    return max(s*x,x)
lrelu=np.vectorize(lrelu)
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(values,lrelu(values),color='red')
ax.set_title('Leaky RELU Function', color='green')
plt.show()
```

## EXPONENTIAL  LINEAR UNIT  FUNCTION (ELU)

The Exponential Linear Unit (ELU) function is a relatively recent modification to the ReLU function. It looks very similar 

to the ReLU function, but it has very different mathematical properties.

\begin{equation}
  {f(x)} =\max{((s{e^{-x}-1)},x)}
\end{equation}


\begin{equation}
  {f(x)} =\begin{cases} x & \text{if $x >= 0 $}\\s(e^-x-1) & \text{otherwise; where s is a small positive value scaler}\end{cases}
\end{equation}

At 0, we don't have a corner. In the case of the ReLU function, we have a corner. In this function, instead of a single 

value going to 0, we have the ELU function slowly going to the negative 's' parameter. It is a good activation function for 

gradient descent optimizers. 
    

```python
import numpy as np
import matplotlib.pyplot as plt
values = np.arange(-5,10,0.1)
s=0.01
def elu(x):
    return max(s*(np.exp(x)-1),x)
elu=np.vectorize(elu)
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(values,elu(values),color='red')
ax.set_title('ELU Function', color='green')
plt.show()
```
